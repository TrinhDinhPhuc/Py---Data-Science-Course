```python
import pandas as pd
import numpy as np
import random 
import matplotlib.pyplot as plt
%matplotlib inline
```

# Explore CO2 emission in VietNam


# Initial the data - Basic plotting

```python
data= pd.read_csv("indicators.csv")
```

```python
data.head()
```

```python
data[data["CountryName"] == "Vietnam"]
```

```python
#set up 2 masks
hist_idicator =  "CO2 emissions \(metric"
hist_country= "VNM"

mask1 = data["IndicatorName"].str.contains(hist_idicator)
mask2 = data["CountryCode"].str.contains(hist_country)

#State is just those indicators matching the USA for country code  an
stage = data[mask1 & mask2]
```

```python
stage.head()
```

```python
# get the years
years = stage["Year"].values
# get the values
co2= stage["Value"].values

# Create a plot
plt.bar(years,co2,)
plt.show
```

here we can obviously see that the emission of Vietnam is increasing dramatically

```python
plt.plot(years,co2)
plt.xlabel("year")
plt.ylabel(stage["IndicatorName"].iloc[2])
plt.title(" CO2 Emission in Vietnam")
```

Overtime, the period of 2000 to 2010 increased rapidly

```python
plt.plot(years,co2)
plt.xlabel('years')
plt.axis([1959,2011,0,5]) #to make more honest, start they y axis at 0
plt.ylabel(stage["IndicatorName"].iloc[2])
plt.title("Co2 emission in Vietnam")
```

# Using histograms to explore the distribution of values
